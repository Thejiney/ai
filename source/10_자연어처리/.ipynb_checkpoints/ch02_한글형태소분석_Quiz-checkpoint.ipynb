# 1. 네이버 api를 활용하여 두 검색어 추출

```python
def get_search_result(query):
    import requests
    import pandas as pd
    import json
    client_id = "XjsU2Un9ix6Pcad3RkZR"
    client_secret = "tm3f2SoXFN"
    words = ''
    for i in range(3):
        url = f'https://openapi.naver.com/v1/search/kin.json?query={query}&display=100&start={i*100+1}'
        headers = {'X-Naver-Client-Id':client_id,
                   'X-Naver-Client-Secret':client_secret}
        response = requests.get(url, headers=headers)
        items = json.loads(response.text)['items']
        for idx, item in enumerate(items):
            description = item['description'].replace('<b>','').replace('</b>','')
            description = description.replace('...',' ')
            words = words + description
    result = words
    return result
```

```python
search1 = get_search_result('경주 여행')
search2 = get_search_result('전주 여행')
```

```python
search1
```

```python
def get_noun(search):
    from konlpy.tag import Kkma
    kkma = Kkma()
    from nltk.tokenize import RegexpTokenizer # 단어 추출(정규표현식을 활용한)
    ret = RegexpTokenizer('[\w]{2,}')
    words = ret.tokenize(search)
    words = ' '.join(words)
    noun_list = [word for word, tag in kkma.pos(words) if tag in ['NNG','NNP']]
    return noun_list
```

```python
noun1 = get_noun(search1)
noun2 = get_noun(search2)
```

```python
import nltk
import matplotlib.pyplot as plt
data_text1 = nltk.Text(noun1)
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(15,4))
data_text1.plot(15)
plt.show()
```

```python
data_text1.dispersion_plot(['코스','추천','곳'])
```

```python
import nltk
import matplotlib.pyplot as plt
data_text2 = nltk.Text(noun2)
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(15,4))
data_text2.plot(15)
plt.show()
```

```python
data_text2.dispersion_plot(['마을','곳','한옥'])
```
