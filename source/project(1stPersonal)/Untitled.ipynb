```python
def get_csv_data(from_date,to_date):
    elem = dv.find_element(By.NAME, 'srhFromDt')
    elem.clear()
    elem.send_keys(from_date.year)
    elem.send_keys('{:02}'.format(from_date.month))
    elem.send_keys('{:02}'.format(from_date.day))
    elem2 = dv.find_element(By.NAME, 'srhToDt')
    elem2.clear()
    elem2.send_keys(to_date.year)
    elem2.send_keys('{:02}'.format(to_date.month))
    elem2.send_keys('{:02}'.format(to_date.day))
    btn_elem = dv.find_elements(By.CLASS_NAME, 'ifdata-search-result')[1]
    btn_elem.click()
    time.sleep(10)
```

```python
while(from_date>start_date):
    get_csv_data(from_date,to_date)
    to_date=from_date
    from_date=to_date-relativedelta(months=1)+relativedelta(days=1)
```

```python
df_only=pd.read_csv('D:/ai/Downloads/Data/apt/행정안전부_지역별(행정동) 성별 연령별 주민등록 1인세대수_20241130.csv', encoding='cp949')
df_only
# 데이터가 너무 세부적으로 분화 되어있음. 시각화에 상당한 어려움이 있음.
```

```python
df_only_index = df_only.loc[...,['시도명','시군구명','읍면동명']]
df_only_values=df_only.iloc[...,8:]
df_only_modify=pd.concat([df_only_index,df_only_values],axis=1)
```

```python
df_groupby=df_only_modify.groupby(['시도명','시군구명']).sum()
```

```python
df_groupby
```

```python
df_groupby_modify = pd.DataFrame(index=df_groupby.index,columns=range(111))
for i in range(len(df_groupby.index)):
    for j in range(110):
        df_groupby_modify.loc[df_groupby_modify.index[i],[j]]=df_groupby.loc[df_groupby.index[i],[f'{j}세남자']][0]+ + df_groupby.loc[df_groupby.index[i],[f'{j}세여자']][0]
```

```python
for i in range(len(df_groupby.index)):
    df_groupby_modify.loc[df_groupby_modify.index[i],[110]]=df_groupby.loc[df_groupby.index[i],['110세이상 남자']][0]+ + df_groupby.loc[df_groupby.index[i],['110세이상 여자']][0]
```

```python
df_groupby_modify
```

```python
df_groupby_modify_age= pd.DataFrame(index=df_groupby.index)

for i in range(len(df_groupby_modify_age.index)):
        df_groupby_modify_age.loc[df_groupby_modify.index[i],'0~5']=df_groupby_modify.loc[df_groupby_modify.index[i],0:5].sum()
        
for i in range(6,110,5):
    for j in range(len(df_groupby_modify_age.index)):
        df_groupby_modify_age.loc[df_groupby_modify.index[j],f'{i}~{i+4}']=df_groupby_modify.loc[df_groupby_modify.index[j],i:i+4].sum()
```
