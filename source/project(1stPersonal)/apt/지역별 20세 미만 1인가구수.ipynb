```python
import pandas as pd
```

```python
df_only_origin = pd.read_csv('D:/ai/Downloads/Data/apt/행정안전부_지역별(행정동) 성별 연령별 주민등록 1인세대수_20241130.csv', encoding='cp949')
df_only_origin
```

```python
df_only=df_only_origin.drop(['행정기관코드','기준연월','계','남자','여자',],axis=1)
df_only=df_only.drop([f'{x}세남자' for x in range(18,110)],axis=1)
df_only=df_only.drop([f'{x}세여자' for x in range(18,110)],axis=1)
df_only=df_only.drop(['110세이상 남자','110세이상 여자'],axis=1)
df_only['18세미만 1인가구수']=df_only.iloc[:,3:].sum(axis=1)
df_only=df_only.drop(list(df_only.columns[3:-1]),axis=1)
df_only=df_only.drop(list(df_only[df_only['18세미만 1인가구수']==0].index),axis=0)
```
