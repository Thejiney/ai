```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import datetime
from dateutil.relativedelta import relativedelta
import os
import pandas as pd
```

```python
dv = webdriver.Chrome()
dv.get('https://rt.molit.go.kr/pt/xls/xls.do?mobileAt=')
```

```python
start_date=datetime.date(2010,1,1)
end_date=datetime.date(2017,1,17)
from_date=end_date-relativedelta(months=1)+relativedelta(days=1)
to_date=end_date
```

```python
def get_csv_data(from_date,to_date):
    elem = dv.find_element(By.NAME, 'srhFromDt')
    elem.clear()
    elem.send_keys(from_date.year)
    elem.send_keys('{:02}'.format(from_date.month))
    elem.send_keys('{:02}'.format(from_date.day))
    elem2 = dv.find_element(By.NAME, 'srhToDt')
    elem2.clear()
    elem2.send_keys(to_date.year)
    elem2.send_keys('{:02}'.format(to_date.month))
    elem2.send_keys('{:02}'.format(to_date.day))
    btn_elem = dv.find_elements(By.CLASS_NAME, 'ifdata-search-result')[1]
    btn_elem.click()
    time.sleep(10)
```

```python
while(from_date>start_date):
    get_csv_data(from_date,to_date)
    to_date=from_date
    from_date=to_date-relativedelta(months=1)+relativedelta(days=1)
```

```python
df_modify=df
df_modify=df_modify.drop(['NO','번지','본번','부번','단지명',\
                          '계약년월','계약일','동','층','매수자','매도자','건축년도','도로명','해제사유발생일',\
                          '거래유형','중개사소재지','등기일자',],axis=1)
df_modify['시도명']=[df['시군구'][x].split(' ')[0] for x in range(len(df.index))]
df_modify['시군구명']=[df['시군구'][x].split(' ')[1] for x in range(len(df.index))]
df_modify['읍면동명']=[df['시군구'][x].split(' ')[2] for x in range(len(df.index))]
df_modify
```

```python
df_modify['거래금액(만원)']=df_modify['거래금액(만원)'].str.replace(',','').astype(int)
df_modify['면적당금액(만원)']=round(df_modify['거래금액(만원)']/df_modify['전용면적(㎡)'])
df_modify=df_modify.drop(['전용면적(㎡)','거래금액(만원)'],axis=1)
df_modify
```

```python
df_modify=round(df_modify.groupby(['시도명','시군구명','읍면동명'],as_index=False).mean())
df_modify
```

```python
df_modify.to_csv('D:/ai/Downloads/Data/apt/실거래가_정리.csv',index=False)
```
