```python
import pandas as pd
```

```python
df_all_origin=pd.read_csv('D:/ai/Downloads/Data/apt/행정안전부_지역별(행정동) 성별 연령별 주민등록 인구수_20241130.csv',encoding='cp949')
df_all_origin
```

```python
df_all=df_all_origin.drop(['행정기관코드','기준연월','계','남자','여자',],axis=1)
df_all=df_all.drop([f'{x}세남자' for x in range(18,110)],axis=1)
df_all=df_all.drop([f'{x}세여자' for x in range(18,110)],axis=1)
df_all=df_all.drop(['110세이상 남자','110세이상 여자'],axis=1)
df_all['18세미만 주민등록 인구수']=df_all.iloc[:,3:].sum(axis=1)
df_all=df_all.drop(list(df_all.columns[3:-1]),axis=1)
df_all=df_all.drop(list(df_all[df_all['18세미만 주민등록 인구수']==0].index),axis=0)
```
